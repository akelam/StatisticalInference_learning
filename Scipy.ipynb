Select at least two distributions, one continuous and one discrete, from the reference of common probability distributions, find their Scipy counterparts via Scipy documentation and plot their histograms. Change the parameters of the distributions and see how they change.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Set up the figure for plotting
fig, ax = plt.subplots(2, 2, figsize=(12, 8))
```

```python
# Normal Distribution (Continuous)
mean1, std_dev1 = 0, 1  # Mean and standard deviation
data_normal1 = stats.norm.rvs(loc=mean1, scale=std_dev1, size=1000)
ax[0, 0].hist(data_normal1, bins=30, density=True, alpha=0.6, color='b')
ax[0, 0].set_title(f'Normal Distribution (mean={mean1}, std_dev={std_dev1})')

# Change parameters for the normal distribution
mean2, std_dev2 = 2, 0.5
data_normal2 = stats.norm.rvs(loc=mean2, scale=std_dev2, size=1000)
ax[0, 1].hist(data_normal2, bins=30, density=True, alpha=0.6, color='g')
ax[0, 1].set_title(f'Normal Distribution (mean={mean2}, std_dev={std_dev2})')

# Poisson Distribution (Discrete)
lambda1 = 3  # Rate parameter (λ)
data_poisson1 = stats.poisson.rvs(mu=lambda1, size=1000)
ax[1, 0].hist(data_poisson1, bins=15, density=True, alpha=0.6, color='r')
ax[1, 0].set_title(f'Poisson Distribution (lambda={lambda1})')

# Change parameters for the Poisson distribution
lambda2 = 7
data_poisson2 = stats.poisson.rvs(mu=lambda2, size=1000)
ax[1, 1].hist(data_poisson2, bins=15, density=True, alpha=0.6, color='m')
ax[1, 1].set_title(f'Poisson Distribution (lambda={lambda2})')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
